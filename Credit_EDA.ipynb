### DATA CLEANING

```python
# The vaues in the DAYS_BIRTH column seems to have an issue as all the values are in negative and as age can't be in negative we need to fix it.
df.DAYS_BIRTH.value_counts()
```

```python
# Removing "-" sign from all the values in DAYS_BIRTH column.

df.DAYS_BIRTH = df.DAYS_BIRTH.apply(lambda x : int(str(x).replace("-", "")))
```

```python
# Checking different education categories present in the Education column.

df.NAME_EDUCATION_TYPE.value_counts()
```

```python
# Converting Secondary/secondary special to Secondary only
df.NAME_EDUCATION_TYPE = df.NAME_EDUCATION_TYPE.apply(lambda x : x[:9] if x == "Secondary / secondary special" else x)
```

```python
# Checking different categories present in the Gender column and at the same time going through their percentage share in the overall data.

df.CODE_GENDER.value_counts(normalize = True)
```

```python
#As we can see there's one category namely XNA and we can easily decide to drop these rows as it's percentage share in the data is rather negligible.

df = df[~(df.CODE_GENDER == "XNA")]
```

```python
# Checking different marital status present in the data set.

df.NAME_FAMILY_STATUS.value_counts()
```

```python
#  By keeping in mind the nature of the dataset we can remove the unknown category as it's presence in the data set is almost negligible and converting Single/ not married to just Single.

df.NAME_FAMILY_STATUS = df.NAME_FAMILY_STATUS.replace("Single / not married", "Single")
df = df[~(df.NAME_FAMILY_STATUS == "Unknown")]
```

```python
# By checking the null values percentage of each column it can be seen the columns from APARTMENTS_AVG to EMERGENCYSTATE_MODE have null values more than 45% so it would be better to drop these columns.

df.drop(df.loc[:, "APARTMENTS_AVG": "EMERGENCYSTATE_MODE"].columns, axis = 1, inplace = True)
```

```python
# Checking for null values again in the data set.

(df.isna().sum()/df.shape[0]*100).sort_values(ascending = False).head(70)
```

```python
# Column OWN_CAR_AGE has 66% of null values and taking nature of data set into account we can take a decision to drop this column.

df.drop("OWN_CAR_AGE", axis = 1, inplace = True)
```

```python
# Converting House / apartment to only House in the NAME_HOUSING_TYPE column.

df.NAME_HOUSING_TYPE = df.NAME_HOUSING_TYPE.replace("House / apartment", "House")
```

```python
# Dropping unnecessary columns i.e. columns with least relevance for performing EDA.

# REGION_POPULATION_RELATIVE
# WEEKDAY_APPR_PROCESS_START
# HOUR_APPR_PROCESS_START

df = df.drop(["HOUR_APPR_PROCESS_START", "WEEKDAY_APPR_PROCESS_START", "REGION_POPULATION_RELATIVE"], axis = 1)
```

```python
# Converting CNT_FAM_MEMBERS values from float to int.

df.CNT_FAM_MEMBERS = df.CNT_FAM_MEMBERS.apply(lambda x : int(x))
```

```python
# Filling null values in AMT_REQ_CREDIT_BUREAU_HOUR,AMT_REQ_CREDIT_BUREAU_DAY,AMT_REQ_CREDIT_BUREAU_WEEK,AMT_REQ_CREDIT_BUREAU_MON,AMT_REQ_CREDIT_BUREAU_QRT,AMT_REQ_CREDIT_BUREAU_YEAR columns with the mode value.

df.AMT_REQ_CREDIT_BUREAU_HOUR.fillna(df.AMT_REQ_CREDIT_BUREAU_HOUR.mode()[0], inplace = True)
df.AMT_REQ_CREDIT_BUREAU_DAY.fillna(df.AMT_REQ_CREDIT_BUREAU_DAY.mode()[0], inplace = True)
df.AMT_REQ_CREDIT_BUREAU_WEEK.fillna(df.AMT_REQ_CREDIT_BUREAU_WEEK.mode()[0], inplace = True)
df.AMT_REQ_CREDIT_BUREAU_MON.fillna(df.AMT_REQ_CREDIT_BUREAU_MON.mode()[0], inplace = True)
df.AMT_REQ_CREDIT_BUREAU_QRT.fillna(df.AMT_REQ_CREDIT_BUREAU_QRT.mode()[0], inplace = True)
df.AMT_REQ_CREDIT_BUREAU_YEAR.fillna(df.AMT_REQ_CREDIT_BUREAU_YEAR.mode()[0], inplace = True)

```

```python
# Converting values in AMT_REQ_CREDIT_BUREAU_HOUR,AMT_REQ_CREDIT_BUREAU_DAY,AMT_REQ_CREDIT_BUREAU_WEEK,AMT_REQ_CREDIT_BUREAU_MON,AMT_REQ_CREDIT_BUREAU_QRT,AMT_REQ_CREDIT_BUREAU_YEAR columns from float to int.


for x in d7:
    df[x] = df[x].apply(lambda y : int(y))
```
